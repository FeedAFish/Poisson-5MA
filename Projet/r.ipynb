```r
source("poisson.r")
library(ggpubr)
suppressMessages(suppressWarnings(library(fanovaGraph)))
library(grid)
library(png)
# library(randtoolbox)
```

```r
expo <- function(lambda,length){
    return(list(X=rexp(length,1/lambda),E_X=lambda))
}
```

```r
li = c <- seq(8,12,.2)
```

```r
resf <- NULL
for (i in seq_along(li)){
    a <- curry(expo,li[i])
    resf <- c(resf,simulate_survival_prob(0,1,100000,.1,a,100000))
}
```

```r
ggplot() + labs(y="Survival probability", x="E[X]") + geom_point(aes(x=li,y=resf))
```

```r
for (i in seq_along(li)){
    png(paste("./plot/f/rplot",toString(li[i]),".png"))
    a <- curry(expo,li[i])
    print(plot_test(0,1,100000,.1,a,5,i,li[i]))
    dev.off()
}
```
